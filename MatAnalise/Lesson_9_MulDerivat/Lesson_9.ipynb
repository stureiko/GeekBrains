–°–µ–≥–æ–¥–Ω—è –Ω–∞ —É—Ä–æ–∫–µ –º—ã –ø—Ä–æ–¥–æ–ª–∂–∏–º –∏–∑—É—á–∞—Ç—å —Ç–µ–º—ã, —Å–≤—è–∑–∞–Ω–Ω—ã–µ —Å —Ñ—É–Ω–∫—Ü–∏—è–º–∏ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö. –í —á–∞—Å—Ç–Ω–æ—Å—Ç–∏:

- —É—Å–ª–æ–≤–Ω—ã–π —ç–∫—Å—Ç—Ä–µ–º—É–º(–º–µ—Ç–æ–¥ –º–Ω–æ–∂–∏—Ç–µ–ª–µ–π –õ–∞–≥—Ä–∞–Ω–∂–∞);
- –º–µ—Ç–æ–¥ –ù—å—é—Ç–æ–Ω–∞;
- –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫;
- –∏–º–∏—Ç–∞—Ü–∏—è –æ—Ç–∂–∏–≥–∞.

### –£—Å–ª–æ–≤–Ω—ã–π —ç–∫—Å—Ç—Ä–µ–º—É–º (–º–µ—Ç–æ–¥ –º–Ω–æ–∂–∏—Ç–µ–ª–µ–π –õ–∞–≥—Ä–∞–Ω–∂–∞)

–í–µ—Ä–Ω–µ–º—Å—è –∫ –ø—Ä–∏–º–µ—Ä—É –∏–∑ –¥–æ–º–∞—à–Ω–µ–≥–æ –∑–∞–¥–∞–Ω–∏—è

–ù–∞–π—Ç–∏ –¥–ª–∏–Ω—É $ùë•$ –∏ —à–∏—Ä–∏–Ω—É $ùë¶$ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–∞ –ø—Ä–∏ –∑–∞–¥–∞–Ω–Ω–æ–º –ø–µ—Ä–∏–º–µ—Ç—Ä–µ $ùëÉ=144$ c–º, –ø—Ä–∏ –∫–æ—Ç–æ—Ä—ã—Ö –¥–∞–Ω–Ω—ã–π –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ –∏–º–µ–µ—Ç –Ω–∞–∏–±–æ–ª—å—à—É—é –ø–ª–æ—â–∞–¥—å $ùëÜ$.

–î–∞–Ω–Ω—É—é –∑–∞–¥–∞—á—É –º—ã —Ä–µ—à–∏–ª–∏ —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ –æ–¥–Ω–æ–≥–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–≥–æ, –ø–æ—Ç–æ–º—É —á—Ç–æ —É—Å–ª–æ–≤–∏–µ –ø–µ—Ä–∏–º–µ—Ç—Ä–∞ –¥–∞–ª–æ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—É—é —Å–≤—è–∑—å –º–µ–∂–¥—É –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º–∏ $ùë•$ –∏ $ùë¶$. –ù–æ –±—ã–≤–∞—é—Ç —Ç–∞–∫–∏–µ —É—Å–ª–æ–≤–∏–µ, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –æ–¥–Ω—É –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –≤—ã—Ä–∞–∑–∏—Ç—å —á–µ—Ä–µ–∑ –¥—Ä—É–≥—É—é –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ. –î–∞–≤–∞–π—Ç–µ —Ä–µ—à–∏–º –Ω–∞—à—É –∑–∞–¥–∞—á—É, –Ω–µ —Å–æ–∫—Ä–∞—â–∞—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö. 

–û—á–µ–≤–∏–¥–Ω–æ, —á—Ç–æ —Ñ—É–Ω–∫—Ü–∏—è, –∫–æ—Ç–æ—Ä—É—é –º—ã –±—É–¥–µ–º –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç—å, —ç—Ç–æ –ø–ª–æ—â–∞–¥—å –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–∞ $S=x\cdot y$. –£—Å–ª–æ–≤–∏–µ –ø–µ—Ä–∏–º–µ—Ç—Ä–∞ –ø–æ–∑–≤–æ–ª—è–µ—Ç –∑–∞–ø–∏—Å–∞—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ: $2\cdot (x+y)=144$, –∏–ª–∏ –ø—Ä–æ—Å—Ç–æ $x+y=72$.

–ü–µ—Ä–≤—ã–º, –∫—Ç–æ —Ä–µ—à–∏–ª –ø–æ–¥–æ–±–Ω—É—é –∑–∞–¥–∞—á—É –±—ã–ª –õ–∞–≥—Ä–∞–Ω–∂, —Å–æ—Å—Ç–∞–≤–∏–≤ —Ñ—É–Ω–∫—Ü–∏—é

### $$L(x,y,\lambda)=x\cdot y +\lambda \cdot (x+y-72)$$

–í –ø–æ—Å–ª–µ–¥—Å—Ç–≤–∏–∏ –æ–Ω–∞ –ø–æ–ª—É—á–∏–ª–∞ –æ–¥–Ω–æ–∏–º–µ–Ω–Ω–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ - __—Ñ—É–Ω–∫—Ü–∏—è –õ–∞–≥—Ä–∞–Ω–∂–∞__.

__–ß–µ–º –∂–µ –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏—è –õ–∞–≥—Ä–∞–Ω–∂–∞ $L(x,y, \lambda)$ –æ—Ç –ø–µ—Ä–≤–æ–Ω–∞—á–∞–ª—å–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –ø–ª–æ—â–∞–¥–∏ $S(x,y)$?__

__1.__ –í –ø–µ—Ä–≤—É—é –æ—á–µ—Ä–µ–¥—å - —ç—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö. –ö–∞–∂–¥–æ–µ –¥–æ–±–∞–≤–ª–µ–Ω–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ –≤ —Ñ—É–Ω–∫—Ü–∏—é –¥–æ–±–∞–≤–ª—è–µ—Ç $\lambda_{i}$. –ù–æ –µ—Å—Ç—å –∏ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É—Å–ª–æ–≤–∏–π –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Å—Ç—Ä–æ–≥–æ –º–µ–Ω—å—à–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –≤ –∏—Å—Å–ª–µ–¥—É–µ–º–æ–π —Ñ—É–Ω–∫—Ü–∏–∏
### $$L(x_{1},x_{2},...,x_{n}, \lambda_{1},\lambda_{2},...,\lambda_{k})=U(x_{1},x_{2},...,x_{n}) +\lambda_{i} \cdot f_{i}(x_{1},x_{2},...,x_{n})$$ 

–≥–¥–µ $i=\overline{1, k}$ –∏ $n>k$

__2.__ –í –ø–ª–∞–Ω–µ –∑–Ω–∞—á–µ–Ω–∏–π —ç—Ç–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π –æ—Å–æ–±–æ –Ω–∏—á–µ–º –æ–Ω–∏ –Ω–µ –æ—Ç–ª–∏—á–∞—Ç—Å—è, —Ç.–∫. –ø–æ —É—Å–ª–æ–≤–∏—é $x+y=72$, —Ç–æ–≥–¥–∞ 
### $$\lambda \cdot (x+y-72)=0$$

–ò—Ç–∞–∫, –∏—Å—Å–ª–µ–¥—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é –Ω–∞ —ç–∫—Å—Ç—Ä–µ–º—É–º.

### $$L(x,y,\lambda)=x\cdot y +\lambda \cdot (x+y-72)$$

–ù–∞–π–¥–µ–º —á–∞—Å—Ç–Ω—ã–µ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ, –ø—Ä–∏—Ä–∞–≤–Ω—è–µ–º –∏—Ö –∫ –Ω—É–ª—é –∏ —Ä–µ—à–∏–º —Å–∏—Å—Ç–µ–º—É —É—Ä–∞–≤–Ω–µ–Ω–∏–π:

### $$\begin{cases}
   L'_{x}=y+\lambda=0, \\
   L'_{y}=x+\lambda=0, \\
   L'_{\lambda}=x+y-72=0
 \end{cases}$$
 
__–ó–∞–º–µ—á–∞–Ω–∏–µ: —á–∞—Å—Ç–Ω–∞—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –ø–æ $\lambda$ —Ä–∞–≤–Ω–∞ —Å–∞–º–æ–º—É —É—Å–ª–æ–≤–∏—é.__ –í—ã—Ä–∞–∑–∏—Ç –∏–∑ –ø–µ—Ä–≤–æ–≥–æ —É—Ä–∞–≤–Ω–µ–Ω–∏—è $y$, –∏–∑ –≤—Ç–æ—Ä–æ–≥–æ $x$ –∏ –ø–æ–¥—Å—Ç–∞–≤–∏–º –≤—Å—ë –≤ —Ç—Ä–µ—Ç—å–µ:

### $$\begin{cases}
   y=-\lambda, \\
   x=-\lambda, \\
   -\lambda-\lambda-72=0
 \end{cases}\Rightarrow 
 \begin{cases}
   y=-\lambda, \\
   x=-\lambda, \\
   -2\lambda=72
 \end{cases}
 \Rightarrow 
 \begin{cases}
   y=36, \\
   x=36, \\
   \lambda=-36
 \end{cases}$$
 
 
–¢.–µ. —Ç–æ—á–∫–∞ $(36,36,-36)$ —è–≤–ª—è–µ—Ç—Å—è —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω–æ–π. –ù–∞–π–¥–µ–º –≤—Ç–æ—Ä—ã–µ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ:
 
### $$L''_{xx}=0$$
### $$L''_{yy}=0$$
### $$L''_{\lambda \lambda}=0$$
### $$L''_{xy}=L''_{yx}=1$$
### $$L''_{x\lambda}=L''_{\lambda x}=1$$
### $$L''_{y\lambda}=L''_{\lambda y}=1$$

__–ó–∞–º–µ—á–∞–Ω–∏–µ: –≤—Ç–æ—Ä–∞—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –ø–æ $\lambda$ –≤—Å–µ–≥–¥–∞ —Ä–∞–≤–Ω–∞ $0$.__ –°–æ—Å—Ç–∞–≤–∏–º –º–∞—Ç—Ä–∏—Ü—É –ì—ë—Å—Å–µ. –î–ª—è —É—Å–ª–æ–≤–Ω–æ–≥–æ —ç–∫—Å—Ç—Ä–µ–º—É–º–∞ –∑–∞–ø–æ–ª–Ω—è–µ–º –µ–µ, –Ω–∞—á–∏–Ω–∞—è —Å $\lambda$.

### $$\begin{pmatrix}
L''_{\lambda \lambda} & L''_{\lambda x} & L''_{\lambda y}\\ 
L''_{x \lambda} & L''_{xx} & L''_{xy}\\
L''_{y \lambda} & L''_{yx} & L''_{yy}\\
\end{pmatrix}\Rightarrow
\begin{pmatrix}
0 & 1 & 1\\ 
1 & 0 & 1\\
1 & 1 & 0\\
\end{pmatrix}$$

–ö —Å–æ–∂–∞–ª–µ–Ω–∏—é, –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ —á–µ—Ä–µ–∑ –º–∏–Ω–æ—Ä—ã –∏–ª–∏ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ, —Ç.–∫. –ø–µ—Ä–≤—ã–π –º–∏–Ω–æ—Ä $\Delta_{1}=0$ –æ–¥–Ω–∞–∫–æ –¥–ª—è —Ñ—É–Ω–∫—Ü–∏–∏ –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –Ω–∞–π—Ç–∏ –≤–µ—Å—å –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å —Ü–µ–ª–∏–∫–æ–º. –ï—Å–ª–∏ $\Delta>0$, –Ω–æ –Ω–∞–π–¥–µ–Ω–Ω–∞—è —Ç–æ—á–∫–∞ –º–∞–∫—Å–∏–º—É–º. –ï—Å–ª–∏ $\Delta<0$, –Ω–æ –Ω–∞–π–¥–µ–Ω–Ω–∞—è —Ç–æ—á–∫–∞ –º–∏–Ω–∏–º—É–º. –ü—Ä–∏ —ç—Ç–æ–º –º–µ—Ç–æ–¥ —á–µ—Ä–µ–∑ –ø–æ–ª–Ω—ã–π –¥–∏—Ñ—Ñ–µ—Ä–µ–Ω—Ü–∏–∞–ª –≤—Ç–æ—Ä–æ–≥–æ –ø–æ—Ä—è–¥–∫–∞ $d^2L$ –ø—Ä–æ–¥–æ–ª–∂–∞–µ—Ç —Ä–∞–±–æ—Ç–∞—Ç—å.

–ù–∞–π–¥–µ–º –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å. –î–ª—è —ç—Ç–æ–≥–æ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —Ä–∞–∑–ª–æ–∂–∏—Ç—å –µ–≥–æ –ø–æ –ª—é–±–æ–π —Å—Ç—Ä–æ–∫–µ –∏–ª–∏ —Å—Ç–æ–ª–±—Ü—É –º–∞—Ç—Ä–∏—Ü—ã –º —É—á–µ—Ç–æ–º –∑–Ω–∞–∫–æ–≤:


### $$\Delta=\begin{vmatrix}
0 & 1 & 1\\ 
1 & 0 & 1\\
1 & 1 & 0\\
\end{vmatrix}=
\begin{vmatrix}
+ & - & +\\ 
- & + & -\\
+ & - & +\\
\end{vmatrix}=0\cdot
\begin{vmatrix}
0 & 1\\
1 & 0\\
\end{vmatrix}-1\cdot
\begin{vmatrix}
1 & 1\\
1 & 0\\
\end{vmatrix}+1\cdot
\begin{vmatrix}
1 & 0\\
1 & 1\\
\end{vmatrix}=(-1)\cdot (-1)+1 \cdot1=2>0$$

–¢.–∫. $\Delta=2>0$, —Ç–æ —Ç–æ—á–∫–∞ $(36,36,-36)$ —è–≤–ª—è–µ—Ç—Å—è –º–∞–∫—Å–∏–º—É–º–æ–º. 

#### –ü—Ä–æ–≤–µ—Ä–∏–º –ø–æ–ª–Ω—ã–π –¥–∏—Ñ—Ñ–µ—Ä–µ–Ω—Ü–∏–∞–ª –≤—Ç–æ—Ä–æ–≥–æ –ø–æ—Ä—è–¥–∫–∞

### $$dL^2=2dxdy+2dxd\lambda+2dyd\lambda$$

–í–æ–∑—å–º–µ–º –Ω–∞—à–µ —É—Å–ª–æ–≤–∏–µ $x+y=72$ –∏ –ø—Ä–æ–¥–∏—Ñ—Ñ–µ—Ä–µ–Ω—Ü–∏—Ä—É–µ–º –æ–±–µ —á–∞—Å—Ç–∏ —Ä–∞–≤–µ–Ω—Å—Ç–≤–∞:

### $$dx+dy=0\Rightarrow dy=-dx$$

–¢–æ–≥–¥–∞ 

### $$dL^2=2dxdy+2dxd\lambda+2dyd\lambda=-2dx^2+2dxd\lambda-2dxd\lambda=-2dx^2$$

–ü—Ä–∏ –ª—é–±—ã—Ö, –æ—Ç–ª–∏—á–Ω—ã—Ö –æ—Ç –Ω—É–ª—è, $dx$ –∏ $dx$ –ø–æ–ª–Ω—ã–π –¥–∏—Ñ—Ñ–µ—Ä–µ–Ω—Ü–∏–∞–ª –≤—Ç–æ—Ä–æ–≥–æ –ø–æ—Ä—è–¥–∫–∞ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã, —ç—Ç–æ —Ç–∞–∫ –∂–µ –≥–æ–≤–æ—Ä–∏—Ç –æ —Ç–æ—á–∫–µ –º–∞–∫—Å–∏–º—É–º–∞.

#### –ü—Ä–æ–≤–µ—Ä–∏–º —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è:

```python
import numpy as np
np.set_printoptions(precision=2, suppress=True)

a = np.array([[0,1,1],[1,0,1],[1,1,0]])
w, v = np.linalg.eig(a)

print(f'–ú–∞—Ç—Ä–∏—Ü–∞ –ì—ë—Å—Å–µ:\n{a}')
print(f'–°–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è:\n{w}')
```

#### –ü—Ä–∏–º–µ—Ä ‚Ññ2
–ò—Å—Å–ª–µ–¥—É–µ–º –Ω–∞ —ç–∫—Å—Ç—Ä–µ–º—É–º —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø–ª–æ—Å–∫–æ—Å—Ç–∏ $U(x,y)=3x+4y-10$ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏, —á—Ç–æ $x^2+y^2=25$

```python
import pylab
from mpl_toolkits.mplot3d import Axes3D
import numpy

fig = pylab.figure()
axes = Axes3D(fig)


def makeData1 ():
    x = numpy.arange (-5, 6, 1)
    U = numpy.arange (140, 340, 20)
    xgrid, Ugrid = numpy.meshgrid(x, U)
    ygrid = numpy.sqrt(25-xgrid**2)
    return xgrid, ygrid, Ugrid
x, y, U = makeData1()
axes.plot_surface(x, y, U, color='b')


def makeData2 ():
    x = numpy.arange (-5, 6, 1)
    U = numpy.arange (140, 340, 20)
    xgrid, Ugrid = numpy.meshgrid(x, U)
    ygrid = -numpy.sqrt(25-xgrid**2)
    return xgrid, ygrid, Ugrid
x, y, U = makeData2()
axes.plot_surface(x, y, U, color='b')

def makeData3 ():
    x = numpy.arange (-10, 11, 1)
    y = numpy.arange (-10, 11, 1)
    xgrid, ygrid = numpy.meshgrid(x, y)
    Ugrid = 3*xgrid+4*ygrid-10
    return xgrid, ygrid, Ugrid
x, y, U = makeData3()
axes.plot_surface(x, y, U, color='r')

pylab.show()
```

–ü—Ä–µ–¥—Å—Ç–∞–≤–∏–º, —á—Ç–æ –º—ã –ø—Ä–∏—à–ª–∏ –Ω–∞ –¥–µ—Ç—Å–∫—É—é –ø–ª–æ—â–∞–¥–∫—É –≤ –ø–µ—Å–æ—á–Ω–∏—Ü—É –≤ —Ñ–æ—Ä–º–µ –Ω–∞—à–µ–π –Ω–∞–∫–ª–æ–Ω–Ω–æ–π –ø–ª–æ—Å–∫–æ—Å—Ç–∏. –í–∑—è–ª–∏ —Ñ–æ—Ä–º–æ—á–∫—É –≤ –≤–∏–¥–µ –Ω–∞—à–µ–≥–æ —É—Å–ª–æ–≤–∏—è $x^2+y^2=25$ - —Ü–∏–ª–∏–Ω–¥—Ä, –∏ –æ–ø—É—Å—Ç–∏–ª–∏ –≤ –ø–µ—Å–æ—á–Ω–∏—Ü—É. –°–∞–º–∞ –ø–æ —Å–µ–±–µ –Ω–∞–∫–ª–æ–Ω–Ω–∞—è –ø–ª–æ—Å–∫–æ—Å—Ç—å —ç–∫—Å—Ç—Ä–µ–º—É–º–∞–º–∏ –Ω–µ –æ–±–ª–∞–¥–∞–µ—Ç. –ù–æ –≤ —Ä–∞–º–∫–∞—Ö —É—Å–ª–æ–≤–∏—è, –º—ã –ø–æ–ª—É—á–∏–º –æ–¥–∏–Ω –º–∏–Ω–∏–º—É–º –∏ –æ–¥–∏–Ω –º–∞–∫—Å–∏–º—É–º. –î–∞–≤–∞–π—Ç–µ –Ω–∞–π–¥–µ–º –∏—Ö.

–°–æ—Å—Ç–∞–≤–∏–º —Ñ—É–Ω–∫—Ü–∏—é –õ–∞–≥—Ä–∞–Ω–∂–∞:

### $$L(x,y,\lambda)=3x+4y-10 +\lambda \cdot (x^2+y^2-25)$$

–ù–∞–π–¥–µ–º —á–∞—Å—Ç–Ω—ã–µ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ, –ø—Ä–∏—Ä–∞–≤–Ω—è–µ–º –∏—Ö –∫ –Ω—É–ª—é –∏ —Ä–µ—à–∏–º —Å–∏—Å—Ç–µ–º—É —É—Ä–∞–≤–Ω–µ–Ω–∏–π:

### $$\begin{cases}
   L'_{x}=3+\lambda \cdot2x=0, \\
   L'_{y}=4+\lambda \cdot2y=0, \\
   L'_{\lambda}=x^2+y^2-25=0
 \end{cases}$$
 
–í—ã—Ä–∞–∑–∏—Ç –∏–∑ –ø–µ—Ä–≤–æ–≥–æ —É—Ä–∞–≤–Ω–µ–Ω–∏—è $y$, –∏–∑ –≤—Ç–æ—Ä–æ–≥–æ $x$ –∏ –ø–æ–¥—Å—Ç–∞–≤–∏–º –≤—Å—ë –≤ —Ç—Ä–µ—Ç—å–µ:
### $$\begin{cases}
   x=-\frac{3}{2\lambda}, \\
   y=-\frac{4}{2\lambda}, \\
   \Bigr(-\frac{3}{2\lambda}\Bigl)^2+\Bigr(-\frac{4}{2\lambda}\Bigl)^2=25
 \end{cases}\Rightarrow 
 \begin{cases}
   x=-\frac{3}{2\lambda}, \\
   y=-\frac{4}{2\lambda}, \\
   \frac{25}{4\lambda^2}=25
 \end{cases}
 \Rightarrow 
 \begin{cases}
   x=-\frac{3}{2\lambda}, \\
   y=-\frac{4}{2\lambda}, \\
   \lambda^2=\frac{1}{4}
 \end{cases}$$
 
–ü–æ–ª—É—á–∞–µ–º –¥–≤–µ —Ç–æ—á–∫–∏: $(-3,-4,\frac{1}{2})$ –∏ $(3,4,-\frac{1}{2})$
 
–ù–∞–π–¥–µ–º –≤—Ç–æ—Ä—ã–µ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ:
### $$L''_{xx}=2\lambda$$
### $$L''_{yy}=2\lambda$$
### $$L''_{\lambda \lambda}=0$$
### $$L''_{xy}=L''_{yx}=0$$
### $$L''_{x\lambda}=L''_{\lambda x}=2x$$
### $$L''_{y\lambda}=L''_{\lambda y}=2y$$ 

–°–æ—Å—Ç–∞–≤–∏–º –º–∞—Ç—Ä–∏—Ü—É –ì—ë—Å—Å–µ:
### $$\begin{pmatrix}
L''_{\lambda \lambda} & L''_{\lambda x} & L''_{\lambda y}\\ 
L''_{x \lambda} & L''_{xx} & L''_{xy}\\
L''_{y \lambda} & L''_{yx} & L''_{yy}\\
\end{pmatrix}\Rightarrow
\begin{pmatrix}
0 & 2x & 2y\\ 
2x & 2\lambda & 0\\
2y & 0 & 2\lambda\\
\end{pmatrix}$$

–î–∞–ª–µ–µ –Ω–∞–π–¥–µ–º –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å –≤ –æ–±—â–µ–º –≤–∏–¥–µ, –∞ –∑–∞—Ç–µ–º —É–∂–µ —Ä–∞–∑–±–µ—Ä–µ–º—Å—è —Å –∫–∞–∂–¥–æ–π —Ç–æ—á–∫–æ–π.

### $$\Delta=\begin{vmatrix}
0 & 2x & 2y\\ 
2x & 2\lambda & 0\\
2y & 0 & 2\lambda\\
\end{vmatrix}=
\begin{vmatrix}
+ & - & +\\ 
- & + & -\\
+ & - & +\\
\end{vmatrix}=0\cdot
\begin{vmatrix}
2\lambda & 0\\
0 & 2\lambda\\
\end{vmatrix}-2x\cdot
\begin{vmatrix}
2x & 0\\
2y & 2\lambda\\
\end{vmatrix}+2y\cdot
\begin{vmatrix}
2x & 2\lambda\\
2y & 0\\
\end{vmatrix}=(-2x)\cdot 4x\lambda+2y \cdot(-4y\lambda)=-8\lambda(x^2+y^2)$$

–ü–æ —É—Å–ª–æ–≤–∏—é –∑–∞–¥–∞—á–∏ $x^2+y^2=25$, —Ç–æ–≥–¥–∞ 
### $$\Delta=-8\lambda \cdot 25 = -200\lambda$$

–¢.–µ. –∑–Ω–∞–∫ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—è –∑–∞–≤–∏—Å–∏—Ç —Ç–æ–ª—å–∫–æ –æ—Ç –∑–Ω–∞–∫–∞ $\lambda$.

–ï—Å–ª–∏ $\lambda=\frac{1}{2}$, —Ç–æ $\Delta<0$, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ $(-3,-4,\frac{1}{2})$ - —Ç–æ—á–∫–∞ –º–∏–Ω–∏–º—É–º–∞.

–ï—Å–ª–∏ $\lambda=-\frac{1}{2}$, —Ç–æ $\Delta>0$, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ $(3,4,-\frac{1}{2})$ - —Ç–æ—á–∫–∞ –º–∞–∫—Å–∏–º—É–º–∞.

### –ú–µ—Ç–æ–¥ –ù—å—é—Ç–æ–Ω–∞

–ï—â–µ –¥–æ —Ä–æ–∂–¥–µ–Ω–∏—è –ñ–æ–∑–µ—Ñ–∞ –õ—É–∏ –õ–∞–≥—Ä–∞–Ω–∂–∞ —Å —Ä–∞–∑–Ω–∏—Ü–µ–π –ø—Ä–∏–º–µ—Ä–Ω–æ –≤ 100 –ª–µ—Ç –∂–∏–ª –Ω–µ –º–µ–Ω–µ–µ –∏–º–µ–Ω–∏—Ç—ã–π —Ñ–∏–∑–∏–∫ –∏ –º–∞—Ç–µ–º–∞—Ç–∏–∫ –ò—Å–∞–∞–∫ –ù—å—é—Ç–æ–Ω.

–í —Ä–∞–º–∫–∞—Ö –¥–∞–Ω–Ω–æ–≥–æ –∫—É—Ä—Å–∞ –Ω–∞–º –∏–Ω—Ç–µ—Ä–µ—Å–Ω—ã –µ–≥–æ —Ç—Ä—É–¥—ã –∫–∞—Å–∞—Ç–µ–ª—å–Ω–æ —Ä–µ—à–µ–Ω–∏—è —Å—Ç–µ–ø–µ–Ω–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π. –ù—å—é—Ç–æ–Ω –ø—Ä–µ–¥–ª–æ–∂–∏–ª —á–∏—Å–ª–µ–Ω–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ —Å—Ç–µ–ø–µ–Ω–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–µ, —Ç–∞–∫ –∂–µ –æ–Ω –æ–±–æ–±—â–∏–ª —Ç–µ–æ—Ä–∏—é –∏ –Ω–∞ —Å–∏—Å—Ç–µ–º—ã —Å—Ç–µ–ø–µ–Ω–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π.

–ü—É—Å—Ç—å $f(x)=0$ - —ç—Ç–æ —Å—Ç–µ–ø–µ–Ω–Ω–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ, —Ç–æ–≥–¥–∞ —Ä–µ—à–µ–Ω–∏–µ –¥–∞–Ω–Ω–æ–≥–æ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –º–æ–∂–Ω–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç—å –≤ –≤–∏–¥–µ $x=\varphi(x)$ –¢.–µ. —Ñ—É–Ω–∫—Ü–∏—è $\varphi(x)$ –¥–æ–ª–∂–Ω–∞ –ø–æ—Å—Ç–µ–ø–µ–Ω–Ω–æ —Å–∂–∏–º–∞—Ç—å—Å—è –≤ —Ö, –ø—Ä–∏ —ç—Ç–æ–º —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏—é $f(x)=0$. –¢.–µ. –µ–µ –º–æ–∂–Ω–æ –∑–∞–ø–∏—Å–∞—Ç—å, –∫–∞–∫:

### $$\varphi(\zeta)=\zeta+\alpha(\zeta) \cdot f(\zeta)$$

–ì–¥–µ –æ—Ç–≤–µ—Ç –Ω–∞ –≤–æ–ø—Ä–æ—Å –±—É–¥–µ—Ç –ª–∏ $\varphi(\zeta)$ —Å–∂–∏–º–∞—Ç—å—Å—è –¥–æ $\zeta$ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç –≤—ã–±–æ—Ä–∞ $\alpha(\zeta)$. –ü–æ—Ç–æ–º—É —á—Ç–æ $f(\zeta)$ –ø—Ä–∏ –æ—á–µ–Ω—å –±–ª–∏–∑–∫–∏—Ö –∑–Ω–∞—á–µ–Ω–∏—è—Ö $\zeta$ –∫ –Ω–∞—à–µ–º—É —Ä–µ—à–µ–Ω–∏—é $x$ –±—É–¥–µ—Ç –±–ª–∏–∑–∫–∞ –∫ $0$.


–î–∞–≤–∞–π—Ç–µ –Ω–∞–π–¥–µ–º –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é:
### $$\varphi'(\zeta)=1+\alpha'(\zeta) \cdot f(\zeta)+\alpha(\zeta) \cdot f'(\zeta)=0$$

–ï—Å–ª–∏ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç—å, —á—Ç–æ $f(\zeta)$ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –±–ª–∏–∑–∫–æ –∫ 0, —Ç–æ –ø–æ–ª—É—á–∏–º:
### $$1+\alpha(\zeta) \cdot f'(\zeta)=0$$

–û—Ç—Å—é–¥–∞
### $$\alpha(\zeta)=-\frac{1}{f'(\zeta)}$$

–û—Å—Ç–∞–ª–æ—Å—å –ø–æ–¥—Å—Ç–∞–≤–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –≤ —Ñ—É–Ω–∫—Ü–∏—é:
### $$\varphi(\zeta)=\zeta-\frac{f(\zeta)}{f'(\zeta)}$$

 –≠—Ç–æ –∏ –µ—Å—Ç—å –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–∏—Ö –º–µ—Ç–æ–¥ –ù—å—é—Ç–æ–Ω–∞, –∫–æ—Ç–æ—Ä—ã—Ö –æ–Ω —Å–∞–º –∏ –ø—Ä–µ–¥–ª–æ–∂–∏–ª.
 
 –ó–∞–º–µ—á–∞–Ω–∏—è:
 
 - —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ–µ —Ä–∞–∑–Ω–æ–æ–±—Ä–∞–∑–∏–µ –º–æ–¥–∏—Ñ–∏–∫–∞—Ü–∏–π –º–µ—Ç–æ–¥–∞ –ù—å—é—Ç–æ–Ω–∞, —Ç.–∫. –≤–æ–ø—Ä–æ—Å –æ –Ω–∞–∏–ª—É—á—à–µ–º –≤—ã–±–æ—Ä–µ $\alpha(\zeta)$ –≤—Å–µ–≥–¥–∞ –æ—Ç–∫—Ä—ã—Ç;
 
 - –ù—å—é—Ç–æ–Ω –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–ª –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ —ç—Ç–æ–≥–æ –º–µ—Ç–æ–¥–∞ –¥–ª—è —Ä–µ—à–µ–Ω–∏—è —Å—Ç–µ–ø–µ–Ω–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π, –∞ —Ç–∞–∫ –∂–µ –∏—Ö —Å–∏—Å—Ç–µ–º. –ù–æ –¥–∞–Ω–Ω—ã–π –º–µ—Ç–æ–¥ —Ç–∞–∫ –∂–µ –ø–æ–¥—Ö–æ–¥–∏—Ç –∏ –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –Ω–µ–ª–∏–Ω–µ–π–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π, –∞ —Ç–∞–∫ –∂–µ –∏—Ö —Å–∏—Å—Ç–µ–º. –ü—Ä–∞–≤–¥–∞ –≤ –ø–æ—Å–ª–µ–¥–Ω–µ–º —Å–ª—É—á–∞–µ –º–µ—Ç–æ–¥ —Å—Ö–æ–¥–∏—Ç—Å—è –Ω–µ –≤—Å–µ–≥–¥–∞ –∏–∑-–∑–∞ –¥–µ–ª–µ–Ω–∏—è –Ω–∞ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é –≤ —Ñ–æ—Ä–º—É–ª–µ. –ò–Ω–æ–≥–¥–∞ –≤ –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª–µ –ø–æ–ª—É—á–∞–µ—Ç—Å—è $0$;
 
 - –º–µ—Ç–æ–¥ –ù—å—é—Ç–æ–Ω–∞ –Ω–∞—Ö–æ–¥–∏—Ç —Ç–æ–ª—å–∫–æ –æ–¥–Ω–æ —Ä–µ—à–µ–Ω–∏–µ, –ø—Ä–∏ —ç—Ç–æ–º –∏—Ö –º–æ–∂–µ—Ç –±—ã—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ. –¢.–µ. –º—ã —Å –ø–æ–º–æ—â—å—é —ç—Ç–æ–≥–æ –º–µ—Ç–æ–¥–∞ –Ω–µ –º–æ–∂–µ—Ç –ø–æ–Ω—è—Ç—å —Å–∫–æ–ª—å–∫–æ –±—ã–ª–æ –∫–æ—Ä–Ω–µ–π –∏ —á–µ–º—É –æ–Ω–∏ –≤—Å–µ —Ä–∞–≤–Ω—ã. 

### –ü—Ä–∏–º–µ—Ä:

–í–æ–∑—å–º–µ–º –¥–≤–µ —Ñ—É–Ω–∫—Ü–∏–∏: 

### $$y_{1}=\sin x$$

### $$y_{2}=-\frac{1}{4}x+\frac{1}{2}$$

```python
from matplotlib import pylab as plt
import numpy as np
%matplotlib inline

plt.figure(figsize=(6,6))
plt.axis([-5, 7, -4, 4])

x=np.linspace(-5, 7, 200)
y1=np.sin(x)
y2=-0.25*x+0.5

plt.plot(x,y1, c='b')
plt.plot(x,y2, c='r')
plt.show()
```

–ü–æ –≥—Ä–∞—Ñ–∏–∫–∞–º –≤–∏–¥–Ω–æ, —á—Ç–æ –æ–Ω–∏ –ø–µ—Ä–µ—Å–µ–∫–∞—é—Ç—Å—è –≤ 3-—Ö —Ç–æ—á–∫–∞—Ö. –ê–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏ —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –Ω–∞–π—Ç–∏ –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ. –î–∞–≤–∞–π—Ç–µ –Ω–∞–π–¥–µ–º –∏—Ö —á–∏—Å–ª–µ–Ω–Ω–æ —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ –ù—å—é—Ç–æ–Ω–∞

```python
from math import sin, cos
from typing import Callable
import unittest

def newton(f: Callable[[float], float], f_prime: Callable[[float], float], x0: float, 
    eps: float=1e-7, kmax: int=1e3) -> float:
    
    x, x_prev, i = x0, x0 + 2 * eps, 0
    
    while abs(x - x_prev) >= eps and i < kmax:
        x, x_prev, i = x - f(x) / f_prime(x), x, i + 1

    return x
# —É—Ä–∞–≤–Ω–µ–Ω–∏–µ sin(x)+x/4-1/2
def f(x: float) -> float:
    return sin(x)+0.25*x-0.5

# –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –æ—Ç sin(x)+x/4-1/2
def f_prime(x: float) -> float:
    return cos(x)+0.25
```

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º —Ä–∞–∑–Ω—ã–µ –Ω–∞—á–∞–ª—å–Ω—ã–µ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏—è:

```python
# –Ω–∞—á–∞–ª—å–Ω–æ–µ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ —Ö0=100
print (newton(f, f_prime, 100))
```

```python
# –Ω–∞—á–∞–ª—å–Ω–æ–µ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ —Ö0=200
print (newton(f, f_prime, 200))
```

```python
# –Ω–∞—á–∞–ª—å–Ω–æ–µ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ —Ö0=300
print (newton(f, f_prime, 300))
```

–ö–∞–∫ –≤–∏–¥–Ω–æ —Ä–µ—à–µ–Ω–∏–µ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç –Ω–∞—á–∞–ª—å–Ω–æ–≥–æ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏—è. –ü—Ä–∏ —ç—Ç–æ–º —Ä–µ—à–µ–Ω–∏–µ –Ω–µ –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –±–ª–∏–∂–µ –≤—Å–µ—Ö –∫ –Ω–∞—á–∞–ª—å–Ω–æ–º—É –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏—é.

–ù–∞ python –Ω–∞–ø–∏—Å–∞–Ω–æ –±–æ–ª—å—à–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ  –±–∏–±–ª–∏–æ—Ç–µ–∫ –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –Ω–µ–ª–∏–Ω–µ–π–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π –∏ –∏—Ö —Å–∏—Å—Ç–µ–º. –í –∫–∞—á–µ—Å—Ç–≤–µ –æ—Å–Ω–æ–≤–Ω–æ–π –≤–æ–∑—å–º–µ–º –±–∏–±–ª–∏–æ—Ç–∫—É scipy.optimize –∏ —Ä–∞–∑–±–µ—Ä–µ–º —Ñ—É–Ω–∫—Ü–∏—é fsolve

```python
from scipy.optimize import fsolve, broyden1
import math

def equations(p):
    x, y = p
    # –ó–∞–ø–∏—Å—å —Å–∏—Å—Ç–µ–º—ã —É—Ä–∞–≤–Ω–µ–Ω–∏—è –≤–∏–¥–∞ x+y=2, x-y=0
    return (x+y-2, x-y)

# –ß–∏—Å–ª–µ–Ω–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ —Å–∏—Å—Ç–µ–º—ã —É—Ä–∞–≤–Ω–µ–Ω–∏–π
x, y =  fsolve(equations, (10, 10))
print (x, y)
```

–ö–∞–∫ –∏ –≥–æ–≤–æ—Ä–∏–ª–æ—Å—å —Ä–∞–Ω–µ–µ, –Ω–µ –≤—Å–µ–≥–¥–∞ –ø–æ–ª—É—á–µ–Ω–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ —è–≤–ª—è–µ—Ç—Å—è –≤–µ—Ä–Ω—ã–º, —Ç.–∫. –∏–Ω–æ–≥–¥–∞ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –¥–µ–ª–µ–Ω–∏–µ –Ω–∞ —á–∏—Å–ª–∞ –±–ª–∏–∑–∫–∏–µ –∫ $0$. –î–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –∂–µ–ª–∞—Ç–µ–ª—å–Ω–æ –≤—ã–≥—Ä—É–∂–∞—Ç—å –≤—Å—é –ø–æ–¥—Ä–æ–±–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é —Å –ø–æ–º–æ—â—å—é –∞—Ç—Ä–∏–±—É—Ç–∞ __full_output=True__. –ù–∞–º –∏–Ω—Ç–µ—Ä–µ—Å–µ–Ω 3 –ø–∞—Ä–∞–º–µ—Ç—Ä, –∫–æ—Ç–æ—Ä—ã–π –æ—Ç–≤–µ—á–∞–µ—Ç –∑–∞ —Ç–æ, –ø—Ä–∏ –∫–∞–∫–∏—Ö —É—Å–ª–æ–≤–∏—è—Ö –±—ã–ª –ø–æ–ª—É—á–µ–Ω –æ—Ç–≤–µ—Ç. –ó–Ω–∞—á–µ–Ω–∏–µ —Ä–∞–≤–Ω–æ–µ $1$ —Å–≤–∏–¥–µ—Ç–µ–ª—å—Å—Ç–≤—É–µ—Ç –æ —Ç–æ–º, —á—Ç–æ –º–µ—Ç–æ–¥ –ø—Ä–æ—à–µ–ª –≤—Å–µ —ç—Ç–∞–ø—ã –∏ —Ä–µ—à–µ–Ω–∏–µ —è–≤–ª—è–µ—Ç—Å—è –≤–µ—Ä–Ω—ã–º.

```python
(x, y), info, ier, mesg =  fsolve(equations, (10, 10), full_output=True)
print (x, y, ier)
```

–ö–∞–∫ –∂–µ –Ω–∞–π—Ç–∏ –≤—Å–µ –∫–æ—Ä–Ω–∏ –Ω–µ–ª–∏–Ω–µ–π–Ω–æ–≥–æ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –º–µ—Ç–æ–¥–æ–º –ù—å—é—Ç–æ–Ω–∞? –î–ª—è —ç—Ç–æ–≥–æ –ø—Ä–∏–¥–µ—Ç—Å—è —Ä–∞–∑–¥—Ä–æ–±–∏—Ç—å —á–∏—Å–ª–æ–≤—É—é –ø—Ä—è–º—É—é –∏ —Ä–µ—à–∏—Ç—å –º–µ—Ç–æ–¥–æ–º –ù—å—é—Ç–æ–Ω–∞ –¥–ª—è –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –Ω–∞—á–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–π. –í–æ–∑—å–º–µ–º –Ω–∞—à—É –ø—Ä–µ–¥—ã–¥—É—â—É—é –∑–∞–¥–∞—á—É —Å –ø–µ—Ä–µ—Å–µ–∫–∞—é—â–∏–º–∏—Å—è –ø—Ä—è–º—ã–º–∏.

### $$\sin x +\frac{1}{4}x-\frac{1}{2}=0$$

```python
import numpy as np
from scipy.optimize import fsolve, broyden2
import math

def equations(p):
    x = p
    return np.sin(x)+0.25*x-0.5

for x in np.arange (-10, 10, 2):
    x, info, ier, mesg =  fsolve(equations, x, full_output=True)
    print (float(x), ier)
```

–ö–∞–∫ –≤–∏–¥–Ω–æ, —á–∞—Å—Ç—å –∫–æ—Ä–Ω–µ–π —è–≤–ª—è–µ—Ç—Å—è –ª–æ–∂–Ω—ã–º–∏, –∏—Ö –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —É–±—Ä–∞—Ç—å. –û—Å—Ç–∞–ª—å–Ω—ã–µ –Ω—É–∂–Ω–æ –Ω–µ–º–Ω–æ–≥–æ –æ–∫—Ä—É–≥–ª–∏—Ç—å –∏ –Ω–µ –∑–∞–ø–∏—Å—ã–≤–∞—Ç—å –¥—É–±–ª–∏. 

```python
import numpy as np
from scipy.optimize import fsolve, broyden2
import math

def equations(p):
    x = p
    return np.sin(x)+0.25*x-0.5

solv = set()
for x in np.arange (-10, 10, 2):
    x, info, ier, mesg =  fsolve(equations, x, full_output=True)
    if (ier == 1): solv.add(int(x*10**8)/10**8)
            
print (solv)
```

–í–æ–∑—å–º–µ–º —Ñ—É–Ω–∫—Ü–∏—é –º–Ω–æ–≥–∏—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö. –î–ª—è –ø—Ä–æ—Å—Ç–æ—Ç—ã –≤–æ—Å–ø—Ä–∏—è—Ç–∏—è –∑–∞–ø–∏—Å–∏ –≤—Å–µ –µ—ë –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –∑–∞–ø–∏—à–µ–º –≤ –≤–∏–¥–µ –≤–µ–∫—Ç–æ—Ä–∞.
### $$U(x_{1},x_{2},...,x_{n})=U(\overrightarrow{x})$$

–¢–æ–≥–¥–∞ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π –º–µ—Ç–æ–¥ –º–æ–∂–Ω–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç—å –≤ —Å–ª–µ–¥—è—â–µ–º –≤–∏–¥–µ:
### $$\overrightarrow{x}_{i+1}=\overrightarrow{x}_i-\alpha(\overrightarrow{x}_i)\cdot \mbox{grad}(U(\overrightarrow{x}_i))$$

–≥–¥–µ $\overrightarrow{x}_i$ - —Ç–µ–∫—É—â–µ–µ –ø–æ–ª–æ–∂–µ–Ω–∏–µ, $\overrightarrow{x}_{i+1}$ - —Å–ª–µ–¥—É—é—â–µ–µ –ø–æ–ª–æ–∂–µ–Ω–∏–µ, $-\mbox{grad}(U(\overrightarrow{x}_i))$ - –∑–Ω–∞—á–µ–Ω–∏–µ –∞–Ω—Ç–∏–≥—Ä–∞–¥–∏–µ–Ω—Ç–∞ –≤ —Ç–æ—á–∫–µ, $\alpha(\overrightarrow{x}_i)$ - –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π —à–∞–≥ –∏–ª–∏ —Å–∫–æ—Ä–æ—Å—Ç—å –¥–≤–∏–∂–µ–Ω–∏—è.

–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–∞—Ä–∏–∞—Ü–∏–π –∑–Ω–∞—á–µ–Ω–∏—è $\alpha(\overrightarrow{x}_i)$ - –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –∏ —Å–∞–º–∏—Ö –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã—Ö –º–µ—Ç–æ–¥–æ–≤ –æ—á–µ–Ω—å –º–Ω–æ–≥–æ.  

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞.

–ë—ã–≤–∞—é—Ç —Å–ª—É—á–∞–∏, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –ø—Ä–∏ —Å–ø—É—Å–∫–µ –º—ã –Ω–µ –¥–æ—Å—Ç–∏–≥–Ω–µ–º –º–∏–Ω–∏–º—É–º–∞, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –≤—Å–µ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–µ —Å–ø—É—Å–∫–∏ –æ–≥—Ä–∞–Ω–∏—á–∏–≤–∞—é—Ç –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –∏—Ç–µ—Ä–∞—Ü–∏–π. –ù–æ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –∫–ª–∞—Å—Å —Ñ—É–Ω–∫—Ü–∏–π, –º–∏–Ω–∏–º—É–º—ã –∫–æ—Ç–æ—Ä—ã—Ö –ø–ª–æ—Ö–æ –∏—â—É—Ç—Å—è —Å –ø–æ–º–æ—â—å –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞. –ù–∞–ø—Ä–∏–º–µ—Ä "–æ–≤—Ä–∞–∂–Ω—ã–µ" —Ñ—É–Ω–∫—Ü–∏–∏.

```python
import pylab
from mpl_toolkits.mplot3d import Axes3D
import numpy

def makeData ():
    x = numpy.arange (-3, 3, 0.01)
    y = numpy.arange (-3, 3, 0.01)
    xgrid, ygrid = numpy.meshgrid(x, y)

    Ugrid = numpy.sin(xgrid*ygrid)
    return xgrid, ygrid, Ugrid

x, y, U = makeData()

fig = pylab.figure()
axes = Axes3D(fig)

axes.plot_surface(x, y, U)

pylab.show()
```

```python
—ç–∫—Å—Ç—Ä–µ–º—É–º–æ–≤import pylab
from mpl_toolkits.mplot3d import Axes3D
import numpy

def makeData ():
    x = numpy.arange (-10, 10, 0.01)
    y = numpy.arange (-10, 10, 0.01)
    xgrid, ygrid = numpy.meshgrid(x, y)

    Ugrid = numpy.sin(xgrid)*numpy.cos(ygrid)
    return xgrid, ygrid, Ugrid

x, y, U = makeData()

fig = pylab.figure()
axes = Axes3D(fig)

axes.plot_surface(x, y, U)

pylab.show()
```

–°–≤–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ –º–µ—Ç–æ–¥ –ø–æ–ª—É—á–∏–ª –ø—Ä–∏ –Ω–∞–±–ª—é–¥–µ–Ω–∏–∏ –∑–∞ –∫—Ä–∏—Å—Ç–∞–ª–ª–∏—á–µ—Å–∫–æ–π —Ä–µ—à–µ—Ç–∫–æ–π –º–µ—Ç–∞–ª–ª–æ–≤ –ø—Ä–∏ –∏—Ö –æ—Ç–∂–∏–≥–µ. –î–µ–ª–æ –≤ —Ç–æ–º, —á—Ç–æ –∫—Ä–∏—Å—Ç–∞–ª–ª–∏—á–µ—Å–∫–∞—è —Ä–µ—à–µ—Ç–∫–∞ –º–æ–∂–µ—Ç –º–µ–Ω—è—Ç—å—Å—è –ø–æ–∫–∞ –≤–µ—Å—å –º–∞—Ç–µ—Ä–∏–∞–ª –Ω–µ –æ—Å—Ç—ã–Ω–µ—Ç –¥–æ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–π —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã.

–ú–µ—Ç–æ–¥ –æ—Ç–∂–∏–≥–∞ –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –≤ —Ç–æ–º, —á—Ç–æ –º—ã –≤–≤–æ–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è —É–±—ã–≤–∞–µ—Ç –¥–æ $0$. –£—Å–ª–æ–≤–Ω–æ –±—É–¥–µ–º –Ω–∞–∑—ã–≤–∞–µ—Ç –µ—ë —Ñ—É–Ω–∫—Ü–∏–µ–π —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã $T(\overrightarrow{x})$. 

–î–∞–ª–µ–µ –±—É–¥–µ–º –ø–µ—Ä–µ–±–∏—Ä–∞—Ç—å —Å–ª—É—á–∞–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º $\overrightarrow{x}$ –∏ —Å—Ä–∞–≤–Ω–∏–≤–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ $U(\overrightarrow{x})$ –≤ —Ç–µ–∫—É—â–µ–π –∏ —Å–ª–µ–¥—É—é—â–µ–π —Ç–æ—á–∫–∏:

- e—Å–ª–∏ –∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –≤ –Ω–æ–≤–æ–π —Ç–æ—á–∫–µ –±–æ–ª—å—à–µ, —á–µ–º –≤ —Ç–µ–∫—É—â–µ–π, —Ç–æ –ø–µ—Ä–µ—Ö–æ–¥–∏–º –≤ –Ω–æ–≤—É—é —Ç–æ—á–∫—É –≤ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é $1$;
- –µ—Å–ª–∏ –º–µ–Ω—å—à–µ, —Ç–æ –ø–µ—Ä–µ—Ö–æ–¥–∏–º —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é, –∫–æ—Ç–æ—Ä–∞—è –ø–æ—Å—Ç–µ–ø–µ–Ω–Ω–æ —É–º–µ–Ω—å—à–∞–µ—Ç—Å—è –∏–∑-–∑–∞ —É–º–µ–Ω—å—à–µ–Ω–∏–µ —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã.

### $$P(\overrightarrow{x}_{i} \to \overrightarrow{x}_{i+1})=
\begin{cases}
   1 & U(\overrightarrow{x}_{i})-U(\overrightarrow{x}_{i+1})<0\\
   e^{\frac{U(\overrightarrow{x}_{i+1})-U(\overrightarrow{x}_{i})}{T(\overrightarrow{x})}} & U(\overrightarrow{x}_{i})-U(\overrightarrow{x}_{i+1})\geqslant0
 \end{cases}$$

## –ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–æ–µ –∑–∞–¥–∞–Ω–∏–µ

__1.__ –ò—Å—Å–ª–µ–¥–æ–≤–∞—Ç—å –Ω–∞ —É—Å–ª–æ–≤–Ω—ã–π —ç–∫—Å—Ç—Ä–µ–º—É–º —Ñ—É–Ω–∫—Ü–∏—é

### $$U=3-8x+6y,$$

–µ—Å–ª–∏

### $$x^2+y^2=36$$

__2.__ –ò—Å—Å–ª–µ–¥–æ–≤–∞—Ç—å –Ω–∞ —É—Å–ª–æ–≤–Ω—ã–π —ç–∫—Å—Ç—Ä–µ–º—É–º —Ñ—É–Ω–∫—Ü–∏—é

### $$U=2x^2+12xy+32y^2+15,$$

–µ—Å–ª–∏

### $$x^2+16y^2=64$$

__3.__ –ß–∏—Å–ª–µ–Ω–Ω–æ –Ω–∞–π—Ç–∏ —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω—ã–π –∫–æ—Ä–µ–Ω—å —Å–∏—Å—Ç–µ–º—ã –Ω–µ–ª–∏–Ω–µ–π–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π:
### $$\begin{cases}
   x^2-y^2+3xy^3-2x^2y^2+2x-3y-5=0 \\
   3y^3-2x^2+2x^3y-5x^2y^2+5=0
 \end{cases}$$

__4*.__ –ß–∏—Å–ª–µ–Ω–Ω–æ –Ω–∞–π—Ç–∏ –≤—Å–µ $5$ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω—ã—Ö –∫–æ—Ä–Ω–µ–π.

__5*.__ –î–∞–Ω—ã –¥–≤–µ —Ñ—É–Ω–∫—Ü–∏–∏ $y_{1}=f(x)$ –∏ $y_{2}=g(x)$. –ò–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ:

### $$f'(x)=\frac{1}{2}g(x)$$

### $$g'(x)=2-2f(x)$$

### $$f(0)=0$$

### $$g(0)=1$$

–í–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å —Ñ—É–Ω–∫—Ü–∏–∏ $y_{1}=f(x)$ –∏ $y_{2}=g(x)$ —Å –ø–æ–º–æ—â—å—é —Ñ–æ—Ä–º—É–ª—ã:

### $$f(x+\Delta x) \approx f(x)+f'(x) \cdot \Delta x$$

–Ω–∞ —É—á–∞—Å—Ç–∫–µ $[0, 20]$

–í –∫–∞—á–µ—Å—Ç–≤–µ —Ä–µ—à–µ–Ω–∏—è –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ —ç—Ç–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π –≤ –æ–¥–Ω–æ–π —Å–∏—Å—Ç–µ–º–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç.


__6**.__ –ù–∞–π—Ç–∏ –≤—Å–µ –∫–æ—Ä–Ω–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏—è:
    
### $$f(x)=g(x)$$

–Ω–∞ —É—á–∞—Å—Ç–∫–µ $[0, 20]$

–ì–¥–µ $y_{1}=f(x)$ –∏ $y_{2}=g(x)$ - —Ñ—É–Ω–∫—Ü–∏–∏ –∏–∑ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ —Ä–µ—à–µ–Ω–∏—è.
