### 6. Исследовать на экстремум функцию: $$U=x^2y+\frac{1}{3}y^3+2x^2+3y^2-1$$

$\displaystyle \begin{cases}
U'_x=2xy+4x=0\\
U'_y=x^2+y^2+6y=0
\end{cases}$

$x,y=(0,0), (\pm\sqrt{8}, -2), (0, -6)$

$\displaystyle \begin{cases}
U''_{xx}=2y+4\\
U''_{yy}=2y+6\\
U''_{xy}=2x
\end{cases} \Rightarrow \begin{vmatrix}
U''_{xx} & U''_{xy}\\ 
U''_{yx} & U''_{yy}\\
\end{vmatrix}=\begin{vmatrix}
2y+4 & 2x\\ 
2x & 2y+6\\
\end{vmatrix}$

$\Delta_{1}=U''_{xx}=2y+4$

$\Delta_{2}=\begin{vmatrix}
U''_{xx} & U''_{xy}\\ 
U''_{yx} & U''_{yy}\\
\end{vmatrix}=U''_{xx}\cdot U''_{yy}-(U''_{xy})^2=(2y+4)(2y+6)-4x^2=4y^2-4x^2+20y+24$

Точка $(0,0): \Delta_1=4>0, \Delta_2 = 24>0 \Rightarrow$ Точка минимума.

Точка $(0,-6): \Delta_1=-8<0, \Delta_2 = 48>0 \Rightarrow$ Точка максимума.

Точка $(\pm\sqrt{8},-2): \Delta_1=0$ седловая точка

```python
from sympy import *
U, x, y = symbols('U x y')
U = x**2*y+1/3*y**3+2*x**2+3*y**2-1
u_x = diff(U,x)
u_y = diff(U,y)
nonlinsolve([u_x, u_y], [x, y])
```

### 7*. Исследовать на экстремум функцию: $$U=e^{-\frac{x}{2}}\cdot(x^2+y^2)$$

$\displaystyle \begin{cases}
U'_x=2x\mbox{e}^{\frac{-x}{2}}-\frac{1}{2}\mbox{e}^{\frac{-x}{2}}(x^2+y^2)=0\\
U'_y=2y\mbox{e}^{\frac{-x}{2}}=0
\end{cases}$

$x,y = (0,0),(4,0)$

$\displaystyle \begin{cases}
U''_{xx}=\mbox{e}^{\frac{-x}{2}}\cdot\Big(\frac{1}{4}x^2-2x+\frac{1}{4}y^2+2\Big)\\
U''_{yy}=2\mbox{e}^{\frac{-x}{2}}\\
U''_{xy}=-y\mbox{e}^{\frac{-x}{2}}
\end{cases} \Rightarrow \begin{vmatrix}
U''_{xx} & U''_{xy}\\ 
U''_{yx} & U''_{yy}\\
\end{vmatrix}$

Точка $(0,0): \Delta_1=2>0, \Delta_2 = 4>0 \Rightarrow$ Точка минимума.

Точка $(4,0): \Delta_1=-\frac{2}{\mbox{e}^2}<0, \Delta_2 = -\frac{4}{\mbox{e}^4}<0 \Rightarrow$ Седловая точка.

```python
U, x, y, e = symbols('U x y e')
U = e**(-x/2)*(x**2+y**2)
u_x = diff(U,x)
u_y = diff(U,y)
u_xx = diff(U, x, x)
u_xy = diff(U, x, y)
u_yy = diff(U, y, y)
D100 = u_xx.subs(x, 0).subs(y, 0)
D200 = u_xx.subs(x, 0).subs(y, 0)*u_yy.subs(x, 0).subs(y, 0)-(u_xy.subs(x, 0).subs(y, 0))**2
D140 = u_xx.subs(x, 4).subs(y, 0)
D240 = u_xx.subs(x, 4).subs(y, 0)*u_yy.subs(x, 4).subs(y, 0)-(u_xy.subs(x, 4).subs(y, 0))**2
```
