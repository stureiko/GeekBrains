```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from statsmodels.graphics.gofplots import qqplot
from scipy.stats import shapiro, normaltest, ttest_rel, ttest_ind, kstest, ks_2samp
```

```python
N = 500
t = np.linspace(-5, 5, N)
xx = np.random.randn(N)
```

```python
plt.hist(xx, label='xx', bins=20)
plt.legend()
plt.show()
```

```python
sns.set_style("white")
sns.kdeplot(xx)
plt.show()
```

```python
sns.distplot(xx)
plt.show()
```

```python
qqplot(xx, line='s')
plt.show()
```

```python
#  -
c, p = shapiro(xx)
print(c, p)
```

```python
#    -  
c, p = normaltest(xx)
print(c, p)
```

```python
#  -
c, p = kstest(xx, 'norm')
print(c, p)
```

```python
yy = np.random.randn(N)+.2
```

```python
sns.kdeplot(xx,alpha=0.5,label='xx')
sns.kdeplot(yy,alpha=0.5,label='yy')
plt.xlabel('t')
plt.ylabel('Density')
plt.title('  xx,yy')
plt.show()
```

```python
#       
c, p = ttest_rel(xx, yy)
print(p)
```

```python
#   - 
c, p = ks_2samp(xx, yy)
print(p)
```
