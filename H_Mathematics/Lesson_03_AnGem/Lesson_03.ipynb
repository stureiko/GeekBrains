```python
from matplotlib.patches import Ellipse
import matplotlib as mpl
import math
from math import sin, cos, pi
```

```python
mean = [ 19.92977907 ,  5.07380955]
width = 30
height = 10.01828848
angle = -54
ell = mpl.patches.Ellipse(xy=mean, width=width, height=height, angle = 180+angle)
fig, ax = plt.subplots()

ax.add_patch(ell)
ax.set_aspect('equal')
ax.autoscale()
plt.show()
```

```python
u=1.     #x-position of the center
v=0.5    #y-position of the center
a=3.     #radius on the x-axis
b=1.5    #radius on the y-axis
t_rot=pi/6 #rotation angle

t = np.linspace(0, 2*pi, 100)
Ell = np.array([a*np.cos(t) , b*np.sin(t)])  
     #u,v removed to keep the same center location
R_rot = np.array([[math.cos(t_rot) , -math.sin(t_rot)],[math.sin(t_rot) , math.cos(t_rot)]])  
     #2-D rotation matrix

Ell_rot = np.zeros((2,Ell.shape[1]))
for i in range(Ell.shape[1]):
    Ell_rot[:,i] = np.dot(R_rot,Ell[:,i])

#plt.plot( u+Ell[0,:] , v+Ell[1,:] )     #initial ellipse
plt.plot( u+Ell_rot[0,:] , v+Ell_rot[1,:])    #rotated ellipse
plt.grid(color='lightgray',linestyle='--')
plt.show()
```

```python
x = np.linspace(0, 6*pi)
y_c = np.cos(x)
y_s = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y_c, x, y_s)
ax.fill_between(x, y_c, y_s, where=y_c>y_s, interpolate=True,
                color='green', alpha=0.3)
ax.fill_between(x, y_c, y_s, where=y_c<y_s, interpolate=True,
                color='blue', alpha=0.3)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
